##Question: Attempt to write a computer program to generate gravity values.

#Answer: See the python code and result below

```python
#import from math library(native to python)
from math import sin, radians




# Function to format latitude into degrees, minutes, and seconds
def format_latitude(total_minutes):
    degrees = total_minutes // 60
    minutes = total_minutes % 60
    seconds = 0  # Always zero since increments are in minutes
    return f"{degrees}° {minutes:02d}′ {seconds:02d}″"  # Use Unicode for prime (′) and double prime (″)


# Function to compute gravity using the International Gravity Formula
def compute_gravity(latitude_degrees):
    # Convert degrees to radians before applying sin()
    latitude_radians = radians(latitude_degrees)
    return 978.0490 * (1 + 0.0052884 * (sin(latitude_radians) ** 2) - 0.0000059 * (sin(2 * latitude_radians) ** 2))


# Print table header
print("{:<15} {:>20}".format("Latitude (° ′ ″)", "Gravity (cm/s²)"))
print("-" * 37)


# Loop through latitude from 0° 0' 0'' to 90° 0' 0'' in increments of 10 minutes
for total_minutes in range(0, 90 * 60 + 1, 10):
    latitude_degrees = total_minutes / 60  # Convert total minutes to degrees
    formatted_latitude = format_latitude(total_minutes)
    gravity = compute_gravity(latitude_degrees)


    # Print formatted output with 6 decimal places
    print("{:<15} {:>20.6f}".format(formatted_latitude, gravity))
```
